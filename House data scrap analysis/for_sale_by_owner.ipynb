```python
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd
import numpy as np
from urllib.request import Request, urlopen
import time
import re
```

```python
class Random_Proxy(object):

    def __init__(self):
        self.__url = 'https://www.sslproxies.org/'
        self.__headers = {
            'Accept-Encoding': 'gzip, deflate, sdch',
            'Accept-Language': 'en-US,en;q=0.8',
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
            'Referer': 'http://www.wikipedia.org/',
            'Connection': 'keep-alive',
            }
        self.random_ip = []
        self.random_port = []

    def __random_proxy(self):

        """
        This is Private Function Client Should not have accesss
        :return: Dictionary object of Random proxy and port number
        """

        r = requests.get(url=self.__url, headers=self.__headers)
        soup = BeautifulSoup(r.text, 'html.parser')

        # Get the Random IP Address
        for x in soup.findAll('td')[::8]:
            self.random_ip.append(x.get_text())

        # Get Their Port
        for y in soup.findAll('td')[1::8]:
            self.random_port.append(y.get_text())

        # Zip together
        z = list(zip(self.random_ip, self.random_port))

        # This will Fetch Random IP Address and corresponding PORT Number
        number = random.randint(0, len(z)-50)
        ip_random = z[number]

        # convert Tuple into String and formart IP and PORT Address
        ip_random_string = "{}:{}".format(ip_random[0],ip_random[1])

        # Create a Proxy
        proxy = {'https':ip_random_string}

        # return Proxy
        return proxy

    def Proxy_Request(self,request_type='get',**kwargs):
        """

        :param request_type: GET, POST, PUT
        :param url: URL from which you want to do webscrapping
        :param kwargs: any other parameter you pass
        :return: Return Response
        """
        while True:
            try:
                proxy = self.__random_proxy()
                return proxy
                break
            except:
                pass
```

```python
num=1
url="https://www.forsalebyowner.com/search/list/california"
#these are the lists the data will be stored in
property_price_storage=[]
property_desc_storage=[]
property_address_storage=[]
property_refrence_storage=[]
property_extractor_storage=[]
```

```python
for i in range(num,45):

    #Pick a random user agent
    #user_agent = random.choice(user_agent_list)

    #this line makes sure the website doesn't block our crawler

    #setting headers
    req = Request(url)

    #Make the request
    webpage = urlopen(req).read()

    #creating soup object
    soup2 = BeautifulSoup(webpage, 'html.parser')

    page_links=set()
    for i in soup2.findAll('a', href=True):
        if "www.forsalebyowner.com/listing" in str(i).lower():
            page_links.add(i['href'])

    for i in page_links:
        url2=i

        req = Request(url2)

        #Make the request
        webpage = urlopen(req).read()

        #creating soup object
        extract_soup = BeautifulSoup(webpage, 'html.parser')



        address=extract_soup.find_all(class_='list-inline list-inline--large')

        for i in address:
            a=(str(i.get_text()))
            a=a.lstrip().rstrip().replace('\n',' ')
            property_address_storage.append(a)


        details=extract_soup.find_all('label',class_='summary-list__label')

        for i in details:
            b=str(i.get_text())
            if '$' in b:
                property_price_storage.append(i.get_text())
            if 'sq' in b:
                area=i.get_text()
            if 'bed' in b:
                beds=i.get_text()
            if 'bath' in b:
                bath=i.get_text()
        property_desc_storage.append(area+' | '+beds+' | '+bath)     


        extractor=extract_soup.find_all('div',class_='cols details')
        extractor_storage=[i.get_text().lstrip().rstrip().replace('\n\n\n','\n').replace('                ','').split('\n') for i in extractor]
        extractor_storage=extractor_storage[0]


        property_extractor_storage.append(extractor_storage)

        property_refrence_storage.append(url2)




        url="https://www.forsalebyowner.com/search/list/california/"+str(num+1)+'-page'
```

```python
detailed_dataframe=pd.DataFrame(list(zip(property_price_storage,
                            property_desc_storage, property_address_storage,property_extractor_storage, property_refrence_storage)),
                                columns=['property_price', 'property_description','property_address','property_details','property_hyperlink'])    

#to construct csv file which includes only houses for sale
detailed_dataframe.to_csv('for_sale_by_owner.csv',index=False)
```
